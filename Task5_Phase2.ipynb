## ðŸŸ¢ Goals Scored vs Conceded by Quarter

```python

import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

quarters = ['1st', '2nd', '3rd', '4th']
scored = [108, 82, 74, 71]
conceded = [54, 52, 49, 55]

x = range(len(quarters))
bar_width = 0.35

fig, ax = plt.subplots(figsize=(10, 6))
bars1 = ax.bar(x, scored, width=bar_width, label='Goals Scored', color='mediumseagreen', edgecolor='black')
bars2 = ax.bar([i + bar_width for i in x], conceded, width=bar_width, label='Goals Conceded', color='coral', edgecolor='black')

# Add text annotations
for bar in bars1 + bars2:
    height = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, height + 1, f'{int(height)}', ha='center', va='bottom', fontsize=9)

ax.set_xticks([i + bar_width / 2 for i in x])
ax.set_xticklabels(quarters)
ax.set_xlabel("Quarter")
ax.set_ylabel("Number of Goals")
ax.set_title("ðŸŸ© Goals Scored vs Conceded by Quarter")
ax.legend()
plt.tight_layout()
plt.show()
```

## ðŸ“‰ Goal Differential by Game

```python

opponents = ['Northwestern', 'Army', 'Maryland', 'Notre Dame', 'Duke', 'Stony Brook', 'Virginia', 'UNC', 'Boston College']
goal_diffs = [-3, 11, -1, 2, 7, -1, 1, 15, -3]

plt.figure(figsize=(12, 6))
sns.lineplot(x=opponents, y=goal_diffs, marker='o')
plt.axhline(0, linestyle='--', color='gray')
plt.title("Goal Differential by Game")
plt.ylabel("Goal Differential")
plt.xlabel("Opponent")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## ðŸŽ¯ Shot Accuracy Comparison â€“ Syracuse vs Opponents

```python

labels = ['Syracuse', 'Opponents']
shots = [716, 552]
shots_on_goal = [522, 367]
accuracy = [sog/s for sog, s in zip(shots_on_goal, shots)]

plt.figure(figsize=(6,4))
sns.barplot(x=labels, y=accuracy, palette='Blues_d')
plt.title("Shot Accuracy Comparison")
plt.ylabel("Accuracy Rate")
plt.ylim(0, 1)
plt.grid(True, axis='y')
plt.tight_layout()
plt.show()
```

## ðŸ§® Goal Sources â€“ Free-Position vs Field Goals

```python

fp_goals = 65
total_goals = 335
field_goals = total_goals - fp_goals

labels = ['Free-Position Goals', 'Field Goals']
values = [fp_goals, field_goals]

plt.figure(figsize=(6,6))
plt.pie(values, labels=labels, autopct='%1.1f%%', startangle=140, colors=['lightsalmon', 'lightblue'])
plt.title("Syracuse Goal Sources â€“ 2024")
plt.axis('equal')
plt.tight_layout()
plt.show()
```
